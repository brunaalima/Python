##Geração de gráficos

```python
import matplotlib.pyplot as plt
import numpy as np
```

####Gráficos em Linha

```python
x= np.linspace(0,2,100)

y1 = x*2
y2 = x**2

plt.plot(x,y1,'*r')
plt.plot(x,y2,'g--')

plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

plt.title('Titulo do Gráfico')
```

```python
x1= np.linspace(0,2,100)

y3 = x*2
y4 = x**2
y5 = x**3
y6 = np.sin(x)

fig, janela = plt.subplots(2,2)

janela[0,0].plot(x1,y3,'r')
janela[0,1].plot(x1,y4,'b')
janela[1,0].plot(x1,y5,'g')
janela[1,1].plot(x1,y6,'y')

fig.legend(['Linear','Quadrática','Cúbica','Senoidal'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1,2,3,4],[10,15,22,31], linestyle='--', color='green', marker='s',linewidth=3.0)

plt.axis([0,10,0,40])
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

###Gráfico em barras


```python
import matplotlib.pyplot as plt
import numpy as np

eixo_y = [10,30,15]
eixo_x = range(len(eixo_y))

largura = 0.5
cor_barra = 'red'
plt.title('Gráfico de barras')

plt.bar(eixo_x,eixo_y,width=largura,color=cor_barra)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1=[1,3,5,7,9]
y1=[5,4,6,8,5]

x2=[2,4,6,8,10]
y2=[3,6,4,9,2]

plt.title('Gráfico de barras comparativo')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.bar(x1,y1, label='Grupo 1')
plt.bar(x2,y2, label='Grupo 2')
plt.legend()
plt.show()
```

###Gráfico de pontos (Scatterplot)

```python
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4]
y=[5,7,2,9]

plt.title('Scatterplot')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.plot(x,y, color='g',linestyle=':')
plt.scatter(x,y, label='Meus pontos', color='k')
plt.legend()
plt.savefig('scatterplot.png', dpi=300)
plt.show()
```

###Histogramas

```python
import matplotlib.pyplot as plt
import numpy as np

num_norm = np.random.randn(1000,1)

plt.hist(num_norm,bins=10)
plt.xlabel("Valores")
plt.ylabel("Frequência Observada")
plt.show()
```
